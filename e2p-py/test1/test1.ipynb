```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

import py360convert


class args:
    i = "demo_equirec.png"
    h_fov = 80
    v_fov = 80
    u_deg = 50
    v_deg = 0
    in_rot_deg = 0
    mode = 'bilinear'
    h = 300
    w = 300
    o = "out.jpg"


# Read image
img = np.array(Image.open(args.i))

# Convert
out = py360convert.e2p(img, fov_deg=(args.h_fov, args.v_fov), u_deg=args.u_deg, v_deg=args.v_deg,
                       out_hw=(args.h, args.w), in_rot_deg=args.in_rot_deg, mode=args.mode)

# Output image
Image.fromarray(out.astype(np.uint8)).save(args.o)

# Show result
plt.imshow(np.array(Image.open(args.o)))
plt.show()
```
